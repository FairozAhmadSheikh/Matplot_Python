```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x=np.arange(1,11,1) #   value of x 
y=x**2  # value of y 
plt.figure(figsize=(4,4)) 
plt.plot(x,y)
plt.title("Graph plotting Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# plotting line with some more features like color , lengend etc 
plt.figure(figsize=(4,4))
plt.plot(x,x**2,c="green",label="x^2")
plt.plot(x,x**3,c='red',label='x^3')
plt.title("Graph plotting two lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
