BAR GRAPH OR BAR PLOT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
students=pd.read_csv('StudentsPerformance.csv')
students
```

```python
# easiest way for bar plotting
level=students['parental level of education'].value_counts().plot(kind='bar',color="orange",figsize=(3,3))
level.set_title("Level of education ")
plt.show()
```
