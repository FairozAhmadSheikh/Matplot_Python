SCATTER PLOT 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
students=pd.read_csv('StudentsPerformance.csv')
students
```

```python
x=students['reading score']
y=students['math score']
plt.figure(figsize=(6,6))
plt.scatter(x,y,alpha=0.5,c='slateblue')
plt.show()
```

```python
# scattering for three features
x=students['reading score']
y=students['math score']
z=students['writing score']

plt.scatter(x,y,c=z,cmap='plasma')
plt.xlabel('Math score')
plt.ylabel('reading score')
plt.colorbar(label='writting score')
plt.show()
```
