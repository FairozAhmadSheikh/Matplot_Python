Plotting with a dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt 

tips=sns.load_dataset('tips')
print(tips.head())
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='time')
plt.show()
```
